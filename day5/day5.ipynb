```python
file1 = open('input.txt', 'r')
lines = file1.readlines()

lines
```

```python
empty_line = 9 # it's 9 in the input, less in the example
```

```python
stacks = lines[0:empty_line]
stacks
```

```python
len(stacks[0])
```

```python
stacks2 = list(zip(*stacks))
stacks2
```

```python
import re

stack_dict = {}

for stack in stacks2:
    if re.search("\d",stack[-1]):
        lst = list(stack[0:-1])
        lst = [x for x in lst if x != ' ']
        stack_dict[stack[-1]] = lst
        
stack_dict
```

```python
instructions = lines[empty_line+1:]
instructions = [line.rstrip() for line in instructions]
instructions
```

```python
quant = []
start_stack = []
end_stack = []

for line in instructions:
    digits = re.findall(r'\d+', line)
    quant.append(digits[0])
    start_stack.append(digits[1])
    end_stack.append(digits[2])
    print(len(quant), len(start_stack), len(end_stack))
```

```python
stack_dict['1']
```

```python
i = 0
start_crate = start_stack[i]
end_crate = end_stack[i]
number_to_move = int(quant[i])
stack_dict[start_crate][0:number_to_move]
stack_dict[start_crate][number_to_move:]
```

```python
stack_dict
```

```python
part1 = True
```

```python
iterations = {}

for i in range(len(quant)):
    start_crate = start_stack[i]
    end_crate = end_stack[i]
    number_to_move = int(quant[i])
    boxes_to_move = stack_dict[start_crate][0:number_to_move].copy()
    stack_dict[start_crate] = stack_dict[start_crate][number_to_move:]
    new_stack = list(boxes_to_move)
    if part1==True:
        new_stack.reverse()
    stack_dict[end_crate] = new_stack + stack_dict[end_crate]
    iterations[i] = stack_dict.copy()
    print(f"Iteration {i}", stack_dict)
```

```python
answer = ''

for crate in stack_dict:
    answer +=stack_dict[crate][0]
    print(answer)
```
