```python
# Using readlines()
file1 = open('input.txt', 'r')
lines = file1.readlines()

lines
```

```python
#lines = [line for line in lines if line == "\n" else line.rstrip("\n")]
lines = [line.rstrip("\n") if line != "\n" else "\n" for line in lines]
lines
```

```python
mystring = ','.join(lines)
```

```python
mystring
```

```python
elves = mystring.split("\n")
elves
```

```python
new_elves = []
for elf in elves:
    new_elf = elf.split(",")
    print(new_elf)
    new_elf = [float(x) if x != "" else 0 for x in new_elf]
    new_elf = sum(new_elf)
    new_elves.append(new_elf)
    
#new_elves
```

```python
max(new_elves)
```
