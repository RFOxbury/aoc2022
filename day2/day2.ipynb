# Part 2 

```python
df2 = df.copy()
```

```python
df2 = df2.rename(columns={"response": "outcome"})
df2.head()
```

```python
df2['outcome'] = df2['outcome'].replace({"rock": "lose", "paper": "draw", "scissors": "win"})
```

```python
df2 = df2[['opponent', 'outcome']]
df2.head()
```

```python
df2['shape'] = np.select(
    [
        (df2['opponent']=='scissors') & (df2['outcome']=='win'), #choose rock=1
        (df2['opponent']=='scissors') & (df2['outcome']=='draw'), #scissors=3
        (df2['opponent']=='scissors') & (df2['outcome']=='lose'), #paper=2
        (df2['opponent']=='rock') & (df2['outcome']=='win'), #paper
        (df2['opponent']=='rock') & (df2['outcome']=='draw'), #rock
        (df2['opponent']=='rock') & (df2['outcome']=='lose'), #scissors
        (df2['opponent']=='paper') & (df2['outcome']=='win'), #scissors
        (df2['opponent']=='paper') & (df2['outcome']=='draw'), #paper
        (df2['opponent']=='paper') & (df2['outcome']=='lose'), #rock
    ], 
    [
        1,
        3,
        2,
        2,
        1,
        3,
        3,
        2,
        1
    ], 
    default=0
)
```

```python
df2['shape'].value_counts()
```

```python
df2.head()
```

```python
df2['score'] = df2['outcome'].replace({"win": 6, "draw": 3, "lose": 0})
df2.head()
```

```python
df2['total'] = df2['shape']+df2['score']
```
