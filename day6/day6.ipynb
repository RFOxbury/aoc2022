```python
file1 = open('input.txt', 'r')
lines = file1.readlines()

lines
```

```python
text = lines[0]
text
```

```python
number_distinct_characters = 4 # 4 for part 1, change to 14 for part 2
```

```python
position_dict = {}
for i in range(0, len(text)):
    position_dict[i+number_distinct_characters] = text[i:i+number_distinct_characters]
```

```python
position_dict
```

```python
for key in position_dict:
    if len(set(position_dict[key]))==number_distinct_characters:
        print(key, position_dict[key])
```
