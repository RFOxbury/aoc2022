# Part 1

```python
totally_contained = []

def test_ranges(range1, range2):
    if (range2[-1] in range1) & (range2[0] in range1):
        return True
    else:
        return False

with open('input.txt') as file:
    for line in file:
        line = line.rstrip()
        line2 = line.split(",")
        line2 = [elf.split("-") for elf in line2]
        line2 = [item for sublist in line2 for item in sublist]
        line2 = [int(number) for number in line2]
        elf1 = range(line2[0],line2[1]+1)
        elf2 = range(line2[2],line2[3]+1)
        if len(elf2) >= len(elf1):
            print("elf 2 has longer range")
            if test_ranges(elf2, elf1)==True:
                totally_contained.append(line2)
        elif len(elf1) >= len(elf2):
            print("elf1 has more numbers")
            if test_ranges(elf1, elf2)==True:
                totally_contained.append(line2)
```

```python
totally_contained
```

```python
len(totally_contained)
```

# Part 2

```python
partial_overlaps = []

def test_ranges_adjusted(range1, range2):
    if (range2[-1] in range1) | (range2[0] in range1):
        return True
    else:
        return False
    
with open('input.txt') as file:
    for line in file:
        line = line.rstrip()
        line2 = line.split(",")
        line2 = [elf.split("-") for elf in line2]
        line2 = [item for sublist in line2 for item in sublist]
        line2 = [int(number) for number in line2]
        elf1 = range(line2[0],line2[1]+1)
        elf2 = range(line2[2],line2[3]+1)
        if len(elf2) >= len(elf1):
            print("elf 2 has longer range")
            if test_ranges_adjusted(elf2, elf1)==True:
                partial_overlaps.append(line2)
        elif len(elf1) >= len(elf2):
            print("elf1 has more numbers")
            if test_ranges_adjusted(elf1, elf2)==True:
                partial_overlaps.append(line2)
```
